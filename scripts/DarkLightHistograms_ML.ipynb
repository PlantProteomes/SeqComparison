**Project Description**

This project reads a given tsv file containing information about some arabidopsis proteomes and make histograms relating count and molecular weight, gravy, and pI statistics of the proteins

```python
# imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

#read TSV file (from Github) into pandas DataFrame
url = "https://raw.githubusercontent.com/PlantProteomes/SeqComparison/main/proteomes/arabidopsis/light_and_dark_protein_list.tsv"
df = pd.read_csv(url, sep="\t")
```

Molecular Weight Overview

```python
# specify right part of data
canonicalProteins = df[df['status']=='canonical']
notObserved = df[df['status']=='not observed']

# find bin number
min = 0
max = 600
binSize = 10
binNum = int((max-min)/binSize)

# plot molecular weight histograms
plt.hist(canonicalProteins['molecular_weight'], label = "Canonical Proteins", alpha = 0.5, edgecolor = "white", bins = binNum, range = [min, max])
plt.hist(notObserved['molecular_weight'], label = "Not Observed", alpha = 0.5, edgecolor = "white", bins = binNum, range = [min, max])

# label and show
plt.xlabel("Molecular Weight")
plt.ylabel("Count")
plt.legend()
plt.xlim(min,max)
plt.show()
```

Zoomed-in Molecular Weight

```python
min = 0
max = 100
binSize = 1
binNum = int((max-min)/binSize)
plt.hist(canonicalProteins['molecular_weight'], label = "Canonical Proteins", alpha = 0.5, edgecolor = "white", bins = binNum, range = [min, max])
plt.hist(notObserved['molecular_weight'], label = "Not Observed", alpha = 0.5, edgecolor = "white", bins = binNum, range = [min, max])

plt.xlabel("Molecular Weight")
plt.ylabel("Count")
plt.legend()
plt.xlim(min,max)
plt.show()
```

```python
min = -2
max = 2
binSize = 0.04
binNum = int((max-min)/binSize)
plt.hist(canonicalProteins['gravy'], label = "Canonical Proteins", alpha = 0.5, edgecolor = "white", bins = binNum, range = [min, max])
plt.hist(notObserved['gravy'], label = "Not Observed", alpha = 0.5, edgecolor = "white", bins = binNum, range = [min, max])

plt.xlabel("Gravy")
plt.ylabel("Count")
plt.legend()
plt.xlim(min,max)
plt.show()
```
